```python
from google.colab import drive
import os
import json
```

```python
# Mount Google Drive
drive.mount("/content/drive", force_remount=True)
```

```python
project_path='/content/drive/MyDrive/Fashion Search Engine Project'
```

```python
# Read the file
file_path =project_path+ '/Data/Lables.json'

with open(file_path, 'r') as file:
    data = json.load(file)
```

```python
fashion_categories = [
    'Suitcase', 'Miniskirt', 'Tie', 'Luggage & bags', 'Shoe', 'Belt', 'Outerwear', 'Dress',
    'Earrings', 'Bracelet', 'Necklace', 'Brassiere', 'Footwear',
    'Satchel', 'Bowtie', 'Top', 'Pants', 'Sunglasses', 'Swimwear', 'Clothing',
    'Glove', 'Skirt', 'High heels', 'Underpants',
    'Fedora', 'Sun hat', 'Sock',
    'Wallet', 'Scarf', 'Watch', 'Umbrella', 'Glasses',
    'Boot', 'Basket', 'Backpack', 'Bag',
    'Hat', 'Coat', 'Sandal', 'Shorts', 'Jeans', 'Shirt', 'Handbag', 'Jacket'
]

# Create a dictionary with sequential IDs
fashion_dict = {item: idx for idx, item in enumerate(fashion_categories)}

# Print the dictionary
for key, value in fashion_dict.items():
    print(f"{key}: {value}")
```

```python
output_directory =project_path+ '/Data/Dataset_Extracted/Images_annotations'
os.makedirs(output_directory, exist_ok=True)
```

```python
# Process and transform the annotations
for image_file, annotations in data.items():
    output_lines = []
    for annotation in annotations:
        label = list(annotation.keys())[0]

        if label in fashion_categories:
            x1, y1, x2, y2 = list(annotation.values())[0]

            x_center = x1 + (x2 - x1) / 2
            y_center = y1 + (y2 - y1) / 2
            width = x2 - x1
            height = y2 - y1
                
            annotation_line = f'{fashion_dict[label]} {x_center} {y_center} {width} {height}\n'
            output_lines.append(annotation_line)

            # Save the transformed annotations to a TXT file
            output_file_path = os.path.join(output_directory, f'{image_file.split(".")[0]}.txt')

            with open(output_file_path, 'w') as file:
                file.writelines(output_lines)

print("Transformation complete. Annotations saved in Google Drive.")
```

```python
# List all files in the folder
file_count = len(os.listdir(output_directory))

print(f"Number of files in the folder: {file_count}")
```
