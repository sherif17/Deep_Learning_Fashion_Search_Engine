```python
import pandas as pd
import json 
```

```python
# Read the Lables file
with open('Lables.json') as file:
    data = json.load(file)

# Extract unique categories
unique_categories = set()
for values in data.values():
    for dictionary in values:
        unique_categories.update(dictionary.keys())

# Count occurrences of each category
category_count = {}
for values in data.values():
    for dictionary in values:
        for category in dictionary.keys():
            category_count[category] = category_count.get(category, 0) + 1
print(f"There are {len(unique_categories)} categories")
print("Unique categories:", unique_categories)
print("Category count:", category_count)
```
